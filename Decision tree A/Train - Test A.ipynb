# Train-Test

```python
%matplotlib inline
import numpy as np
from pylab import *
import matplotlib.pyplot as plt

np.random.seed(2)

pageSpeeds = np.random.normal(3.0, 1.0, 100)
purchaseAmount = np.random.normal(60.0, 40.0, 100) / pageSpeeds


scatter(pageSpeeds, purchaseAmount)
```

```python
trainX = pageSpeeds[:80]
testX = pageSpeeds[80:]

trainY = purchaseAmount[20:]
testY = purchaseAmount[:20]
```

```python
scatter(trainX, trainY)
```

```python
scatter(testX, testY)
```

```python
x = np.array(trainX)
y = np.array(trainY)

p4 = np.poly1d(np.polyfit(x, y, 3))
```

```python
xp = np.linspace(0, 6, 100)
axes = plt.axes()
axes.set_xlim([0,6])
axes.set_ylim([0, 200])
plt.scatter(x, y)
plt.plot(xp, p4(xp), c='r')
plt.show()
```

```python
testx = np.array(testX)
testy = np.array(testY)

axes = plt.axes()
axes.set_xlim([0,7])
axes.set_ylim([0, 200])
plt.scatter(testx, testy)
plt.plot(xp, p4(xp), c='g')
plt.show()
```

```python
from sklearn.metrics import r2_score

r2 = r2_score(testy, p4(testx))

print(r2)
```

```python
r2 = r2_score(np.array(trainY), p4(np.array(trainX)))
print(r2)
```
